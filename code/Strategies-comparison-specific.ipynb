```python
import pandas as pd
import seaborn as sb
import os
import numpy as np
```

```python

cols = ["var", "vsa", "vca", "initTime", "solveTime", "solutions", "variables", "propagators", "propagations", "nodes", "failures", "restarts", "peakDepth"]

model = open("v03search2.mzn",'r')

base = "solve :: int_search( "

#cmd = 'minizinc searchV01.mzn -s'
#print(cmd)
#out = os.popen(cmd).read()
stringa_file = model.read().split("%*")
model.close()
#print(stringa_file[1])
```

```python
results = pd.DataFrame(columns=cols);
```

```python
x = 5

var2      = ["P", "X", "scheduling"]
vsa2 = ["dom_w_deg", "first_fail", "impact", "input_order", "largest", "max_regret", "most_constrained", "smallest"]
# "occurrence",

vca2 = ["indomain", "indomain_max",  "indomain_min","indomain_random","indomain_reverse_split","indomain_split","indomain_split_random"]
# "indomain_median",
```

```python
# all the others

for i in var2:
    print(i)
    for j in vsa2:
        for h in vca2:
            for l in range(x):
                
                nuovo_modello = stringa_file[0] + "%*" + os.linesep + base + i + ", " + j + ", " + h + ") minimize sum(total_travel);" + os.linesep +"%*" + stringa_file[2]

                f = open("v03search2.mzn", "w")
                f.write(nuovo_modello)
                f.close()

                cmd = "minizinc --solver Gecode v03search2.mzn data/Visano_dijkstra.dzn data/temp-order-to-test.dzn --statistics -t 20000"

                out = os.popen(cmd).read()
                temp = [line for line in out.split("\n") if "%%%mzn-stat: " in line]
                print(temp)

                stats = {}
                for k in range(3, len(cols)):
                    if [float(line.split("=")[1]) for line in temp if "%%%mzn-stat: " + cols[k] in line] != []:
                        stats[cols[k]] = [float(line.split("=")[1]) for line in temp if "%%%mzn-stat: " + cols[k] in line][0]
                    else:
                        stats[cols[k]] = np.NaN

                row = {"var": i, "vsa": j, "vca": h}
                row.update(stats)

                results = results.append(row, ignore_index = True)

                results.to_csv("resultsV03-specific.csv")
```
