# Analisi dei dati su alcune delle strategie ripetute
- Prima di fare qualsiasi cosa vengono eliminate le colonne che non ci danno nessuna informazione perchè contengono sempre lo stesso dato.
- Poi vengono eliminate tutte le righe che come tempo di esecuzione hanno un valore nullo, ciò implica che l'esecuzione ha raggiunto il timeout e non ha stampa alcun valore per la statistica
- Infine vengono rimessi a posto gli indici della tabella

```python
d = pd.read_csv('resultsV03-specific.csv')
df = d[1:]
df = df.drop(columns=['Unnamed: 0', 'variables', 'propagators', 'restarts'])
df = df[df['solveTime'] >= 0 ].copy()
df = df.reset_index()
df = df.drop(columns=['index'])
df
```

```python
corr = df.corr()
f, ax = plt.subplots(figsize=(18,18))
cmap = sb.diverging_palette(250, 20, as_cmap=True)
xx = sb.heatmap(corr.round(2), cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
```

### Analisi matrice di correlazione
La matrice di correlazione ci mostra come quasi tutte le variabili analizzate siano legate tra loro, in particolare:
- il numero di fallimenti e il numero di nodi(1)
- il numero di propagazioni e il solveTime(0.79), anche analizzando la profondità massima e il solveTime(-0.05), ci indica che all'aumentare del numero di propagazioni aumenta anche il tempo per completare la ricerca ma non aumenta la profondità della'albero. Questo ci fa pensare al fatto che avere alberi larghi rispetto che profondi aumenti il tempo di esecuzione genereale(???? ipotesi)
